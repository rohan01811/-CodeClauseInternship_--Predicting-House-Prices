```python
from sklearn.model_selection import train_test_split as tts
import  sklearn.linear_model as lm
```

```python
df.head()
```

```python
x=df.iloc[:,[1,2]]
```

```python
x
```

```python
y=df["price"]
```

```python
y
```

```python
x_train,x_test,y_train,y_test = tts(x,y,test_size = 0.2)
```

```python
x_train
```

```python
y_train
```

```python
model=lm.LinearRegression()
```

```python
x1_mul = np.array(x_train)
y1_mul = np.array(y_train)
x2_mul = np.array(x_test)
y2_mul = np.array(y_test)
np.shape(x1_mul)
```

```python
x1_mul
```

```python
y1_mul
```

```python
x1_mul.shape
```

```python
y1_mul.shape
```

```python
model.fit(x1_mul,y1_mul)
```

```python
y_pred = model.predict(x2_mul)
y_pred.shape
```

```python
y_pred
```

```python
from sklearn.metrics import mean_squared_error 
```

```python
meanerr2=mean_squared_error(y2_mul,y_pred)
```
